```python
IARC:
    A Agência Internacional de Pesquisa em Câncer (Iarc, na sigla em inglês) é a agência especializada em câncer da Organização Mundial ds Saúde (OMS), estabelecida pela resolução WHA18.44 na 18ª Assembleia Mundial da Saúde, em 20 de maio de 1965.

A Iarc tem como objetivo promover a colaboração internacional na pesquisa em câncer sempre com foco na prevenção da doença. Sua função principal consiste em realizar e coordenar a pesquisa e investigação científica em nível internacional, além de auxiliar na implementação de politicas eficazes para prevenção do câncer.  A pesquisa oferece evidência científica forte e independente para a criação de diretrizes e políticas de controle da doença pelas autoridades internacionais e nacionais.

Além da sua excelência na pesquisa científica, a Agência também é responsável por produzir publicações de alta relevância global sobre câncer. Dentre elas, destacam-se Cancer Incidence in Five Continents e Globocan, de alta relevância sobre estatística mundial de câncer;  Iarc Monographs on the Evaluation of Carcinogenic Risks to Humans (avaliação de risco de agentes cancerígenos para humanos);  Iarc Handbooks of Cancer Prevention (avaliação de estratégias de prevenção do câncer); e WHO Classification of Tumours (classificação anatomopatológica do câncer).

A Iarc tem atualmente 26 Estados Participantes que apoiam e supervisionam o trabalho da Agência, atuando nas suas estruturas de governança: os conselhos Científico e Diretivo. 

    
    
https://www.otempo.com.br/interessa/aumento-do-cancer-no-brasil-e-na-america-latina-preocupa-1.2128557
https://www.iarc.fr/about-iarc-membership/
###O Brasil é membro da Iarc desde 2013, sendo o único país da América Latina. O INCA, representando o Brasil, possui assento tanto no Conselho Diretivo, votado às políticas e decisões da Agência, como no Conselho Científico, marcando posição na estratégia científica da Iarc, e com isso garantindo que sejam consideradas as necessidades de investigação em domínios de especial relevância para o País e para a região sul-americana.
```
